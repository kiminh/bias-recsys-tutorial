## Item popularity in the dataset 

```python
dataset = 'ml1m'
method = 'utime'
mode = 'implicit'
user_field = 'user_id'
item_field = 'item_id'
rating_field = 'rating'
type_field = 'type_id'
cutoffs = '5,10,20,50,100,200'
```

```python
data =  pd.read_csv('../data/outputs/splits/' + dataset + '_' + method + '.csv', encoding='utf8')
```

```python
item_pop = data.groupby([item_field]).count().sort_values(user_field, ascending=False)[user_field]
```

```python
head_tail_split = 808
print('Head Tail', compute_gini(item_pop[:head_tail_split]))
print('Long Tail', compute_gini(item_pop[head_tail_split:]))
head_tail_items = np.array(item_pop[:head_tail_split].index)
long_tail_items = np.array(item_pop[head_tail_split:].index)
```

```python
plt.rcParams.update({'font.size': 8})
plt.figure(figsize=(6, 3.5))

plt.title(r'Item popularity')
plt.xlabel('Items ranked by popularity')
plt.ylabel('Number of ratings in the dataset')
plt.plot(range(head_tail_split), item_pop.values[:head_tail_split], alpha=0.7, label=r'Head tail')
plt.plot(range(head_tail_split, len(item_pop.index)), item_pop.values[head_tail_split:], label=r'Long tail')
plt.axhline(y=item_pop.values[head_tail_split], linestyle='--', lw=1, c='grey')
plt.axvline(x=head_tail_split, linestyle='--', lw=1, c='grey')
plt.xlim([-25, len(item_pop.index)])
plt.ylim([-25, item_pop.values[0]])
plt.legend()
plt.tight_layout()
```

## Data analysis on recommendation algorithm metrics

```python
l_cutoffs = np.array([int(k) for k in cutoffs.split(',')])[:2]
```

```python
model_types = ['utime_bpr_xquad', 'utime_bpr', 'utime_random', 'utime_top_popular']
```

```python
metrics = {}
for model_type in model_types:
    metrics[model_type] = load_obj(os.path.join(data_path, 'outputs/metrics/' + dataset + '_' + model_type + '_metr.h5'))
```

```python
plt.rcParams.update({'font.size': 16.5})
plt.figure(figsize=(30, 7.5))

plt.subplot(131)
plt.title(r'Precision')
plt.xlabel('Cutoff Value')
plt.ylabel('Precision')
for model_type in model_types:
    plt.plot(l_cutoffs, [np.mean(metrics[model_type]['precision'][k,:]) for k in range(len(l_cutoffs))], label=model_type)
plt.xticks(l_cutoffs)
plt.legend()
plt.grid(axis='y')

plt.subplot(132)
plt.title(r'Recall')
plt.xlabel('Cutoff Value')
plt.ylabel('Recall')
for model_type in model_types:
    plt.plot(l_cutoffs, [np.mean(metrics[model_type]['recall'][k,:]) for k in range(len(l_cutoffs))], label=model_type)
plt.xticks(l_cutoffs)
plt.legend()
plt.grid(axis='y')

plt.subplot(133)
plt.title(r'NDCG')
plt.xlabel('Cutoff Value')
plt.ylabel('NDCG')
for model_type in model_types:
    plt.plot(l_cutoffs, [np.mean(metrics[model_type]['ndcg'][k,:]) for k in range(len(l_cutoffs))], label=model_type)
plt.xticks(l_cutoffs)
plt.legend()
plt.grid(axis='y')

plt.tight_layout()
```

```python
plt.rcParams.update({'font.size': 16.5})
plt.figure(figsize=(30, 7.5))

plt.subplot(131)
plt.title(r'Average Popularity of Items')
plt.xlabel('Cutoff Value')
plt.ylabel('APT')
for model_type in model_types:
    plt.plot(l_cutoffs, [np.mean(metrics[model_type]['mean_popularity'][k,:]) for k in range(len(l_cutoffs))], label=model_type)
plt.legend()

plt.subplot(132)
plt.title(r'Average Percentage of Long-Tail Items')
plt.xlabel('Cutoff Value')
plt.ylabel('APLT')
for model_type in model_types:
    plt.plot(l_cutoffs, [np.sum(metrics[model_type]['item_coverage'][k,long_tail_items]) / np.sum(metrics[model_type]['item_coverage'][k,:]) for k in range(len(l_cutoffs))], label=model_type)
plt.legend()

plt.subplot(133)
plt.title(r'Item Coverage')
plt.xlabel('Cutoff Value')
plt.ylabel('Item Coverage')
for i, model_type in enumerate(model_types):
    plt.plot(l_cutoffs, [len([1 for m in metrics[model_type]['item_coverage'][k,head_tail_items] if m > 0]) / len(head_tail_items) for k in range(len(l_cutoffs))], color='C'+str(i), linestyle='-', label='head tail of ' + model_type)
    plt.plot(l_cutoffs, [len([1 for m in metrics[model_type]['item_coverage'][k,long_tail_items] if m > 0]) / len(long_tail_items) for k in range(len(l_cutoffs))], color='C'+str(i), linestyle=':', label='long tail of ' + model_type)
plt.legend()

plt.tight_layout()
```

## Sample treatment for popularity

#### Prepare data

```python
traintest = pd.read_csv('../data/outputs/splits/' + dataset + '_' + method + '.csv', encoding='utf8')
train = traintest[traintest['set']=='train'].copy()
test = traintest[traintest['set']=='test'].copy()
```

```python
users = list(np.unique(traintest[user_field].values))
items = list(np.unique(traintest[item_field].values))
```

```python
items_metadata = traintest.drop_duplicates(subset=['item_id'], keep='first')
category_per_item = items_metadata[type_field].values
```

```python
if mode == 'implicit':
    train[rating_field] = train[rating_field].apply(lambda x: 1.0)
    test[rating_field] = test[rating_field].apply(lambda x: 1.0)
    traintest[rating_field] = traintest[rating_field].apply(lambda x: 1.0)
```

#### Setup and use a reranking model

```python
from models.ranker_xquad import RankerXQuad
original_model_type = 'utime_bpr'
reranked_model_type = original_model_type + '_' + 'xquad'
model = RankerXQuad(users, items, train, test, category_per_item, item_field, user_field, rating_field)
```

```python
predictions = load_obj(os.path.join(data_path, 'outputs/predictions/' + dataset + '_' + original_model_type + '_pred.h5'))
```

```python
model.set_predictions(predictions)
```

```python
model.rerank(type='smooth', lmbda=0.4, k=10, rmax=100, head_tail_split=head_tail_split)
```

```python
save_obj(model.get_predictions(), os.path.join(data_path, 'outputs/predictions/' + dataset + '_' + reranked_model_type + '_pred.h5'))
```

```python
model.test(cutoffs=np.array([int(k) for k in cutoffs.split(',')]))
```

```python
save_obj(model.get_metrics(), os.path.join(data_path, 'outputs/metrics/' + dataset + '_' + reranked_model_type + '_metr.h5'))
```
